```python
import os
import pathlib
import json
import numpy as np
```

```python
prod_dir = pathlib.Path("prod")
data = []
for d_file in [nm for nm in os.listdir(prod_dir) if nm.endswith(".json")]:
    try:
        with open(prod_dir/d_file,"r",encoding="utf-8") as f:
            js = json.load(fp=f)
            data.append((d_file.rstrip(".json"),np.asarray(js)))
    except Exception as e:
        e
```

```python
import matplotlib.pyplot as plt

# plt.style.use('_mpl-gallery-nogrid')
fig, ax = plt.subplots(len(data))
w=20
fig.set_size_inches(w*2,w)

if len(data)==1:
  ax.imshow(data[0][1], origin='upper')
else:
  for i,dt in enumerate(data):
    ax[i].imshow(dt[1], origin='upper')
    ax[i].set_title(dt[0])


plt.show()
```
